```python
import matplotlib.pyplot as plt
```

```python
# AlexNet on ImageNet singlecrop top1 accuracy in %
configs = ["32/32/32\n(Baseline)", "8/8/8", "2/2/32", "1/2/32", "1/2/6"]
accuracy = [55.9, 53.0, 50.3, 47.7, 46.1]

plt.style.use("ggplot")
plt.figure(figsize=(8, 4), dpi=130)
bars = plt.bar(configs, accuracy)
plt.ylabel("Top-1 Accuracy (%)")
plt.xlabel("Configuration (Weights/Activations/Gradients)")
plt.title("DoReFa-Net Accuracy with Bit-Width Configuration")

for bar, acc in zip(bars, accuracy):
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.4,
        f"{acc:.1f}%",
        ha="center",
        va="bottom",
    )

plt.ylim(0, 60)
plt.tight_layout()
plt.show()
```
